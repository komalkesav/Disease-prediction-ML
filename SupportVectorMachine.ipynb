```python
# Import required libraries
import numpy as np
import pandas as pd
from sklearn import svm
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score,confusion_matrix
```

```python
data_train = pd.read_csv('Training.csv')
data_test = pd.read_csv('Testing.csv')
```

```python
data_train.head()
```

```python
data_test.head()
```

```python
x_test= data_test.iloc [:, : 132]
y_test= data_test.iloc [:, 132]
```

```python
x_train= data_train.iloc [:, : 132]
y_train= data_train.iloc [:, 132]
```

```python
# Create SVM model
clf = svm.SVC(kernel='linear')
```

```python
# Train SVM model
clf.fit(x_train, y_train)
```

```python
# Make predictions on test data
y_pred = clf.predict(x_test)
```

```python
# Evaluate model accuracy
cm_svm = confusion_matrix(y_test, y_pred)
accuracy_svm = accuracy_score(y_test, y_pred)
precision_svm = precision_score(y_test, y_pred, average='weighted')
recall_svm = recall_score(y_test, y_pred, average='weighted')
f1_svm = f1_score(y_test, y_pred, average='weighted')
```

```python
print("Confusion Matrix:", cm_svm)
print("Accuracy:", accuracy_svm)
print("SVM Precision:", precision_svm)
print("SVM Recall:", recall_svm)
print("SVM F1 Score:", f1_svm)
```
